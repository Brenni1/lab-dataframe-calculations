## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
orders.nunique()
```

```python
sum_spent = orders.groupby(['customerid']).sum() #quick and dirty way to get the total amount spent per customer
sum_spent
```

```python
sum_spent_95 = np.percentile(sum_spent['amount_spent'],95) #calculating the 95 percentile
sum_spent_95
```

```python
sum_spent_75 =  np.percentile(sum_spent['amount_spent'],75) #calculating the 75 percentile
sum_spent_75
```

```python
def get_customer(x):            # function to assign the correlating status to each customer
    if x > sum_spent_95:
        return 'VIP'

    if x > sum_spent_75  and x <= sum_spent_95:
        return 'Preferred'

    else:
        return 'Other'
```

```python
sum_spent['customer_type'] = sum_spent['amount_spent'].apply(get_customer)  #applying the function to our summed list
sum_spent.head()
```

```python
sum_spent['customer_type'].value_counts() #checking for the amount of customertypes
```

## Q2: How to identify which country has the most VIP Customers?

```python
country_agg = orders.groupby(['customerid','country']).agg({'amount_spent':sum})  #creating a clean aggregated list with only relevant information
country_agg.head()
```

```python
country_agg['customer_type'] = country_agg['amount_spent'].apply(get_customer) # apply the function to get the customer_type
country_agg.head(10)
```

```python
vip = country_agg[country_agg['customer_type']=='VIP'].reset_index() #filtering for only vips and resetting the index to be able to filter for country in the next step
vip.head()
```

```python
vip['country'].value_counts()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
vippref = country_agg[country_agg['customer_type']!='Other'].reset_index() #prefiltering for only vip+preferred customers
vippref.head(10)
```

```python
vippref['country'].value_counts()
```
